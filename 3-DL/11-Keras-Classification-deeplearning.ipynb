In the tutorial we learned about the building blocks of neural networks: *linear units*. Before you get started, run the code cell below to set everything up.

The *Red Wine Quality* dataset consists of physiochemical measurements from about 1600 Portuguese red wines.  Also included is a quality rating for each wine from blind taste-tests.

First, run the next cell to display the first few rows of this dataset.

```python
import pandas as pd
from IPython.display import display
```

```python
red_wine = pd.read_csv('winequality-red.csv')
```

```python
red_wine.index
```

```python
# Create training and validation splits
df_train = red_wine.sample(frac=0.7, random_state=0)
```

```python
df_train.index
```

```python
df_train.describe()
```

```python
df_valid = red_wine.drop(df_train.index)
```

```python
df_valid.describe()
```

```python
display(df_train.head(4))
```

```python
df_train.describe()
```

```python
# Scale to [0, 1]
max_ = df_train.max(axis=0)
min_ = df_train.min(axis=0)
```

```python
max_
```

```python
min_
```

```python
df_train = (df_train - min_) / (max_ - min_)
df_valid = (df_valid - min_) / (max_ - min_)
```

```python
df_train.describe()
```

```python
df_valid.describe()
```

```python
df_train["quality"].min()
```

```python
# Split features and target
X_train = df_train.drop('quality', axis=1)
X_valid = df_valid.drop('quality', axis=1)
y_train = df_train['quality']
y_valid = df_valid['quality']
```

```python
X_train.shape
```

```python
X_train.describe()
```

```python
y_train.describe()
```

```python
X_valid.describe()
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=[11]),
    layers.Dense(512, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(12, activation='relu'),
    layers.Dense(1, activation='sigmoid'),
])
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import SGD
```

```python
opt = SGD(learning_rate=0.01, momentum=0.5)

model.compile(
    optimizer='adam',
    loss='binary_crossentropy', metrics= ['accuracy']
)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    batch_size=256,
    epochs=200,
)
```
