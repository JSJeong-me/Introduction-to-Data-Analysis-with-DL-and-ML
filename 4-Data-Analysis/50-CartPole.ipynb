```python
import gym
import numpy as np
from PIL import Image

def discretize(obs):
    env_low = env.observation_space.low
    env_high = env.observation_space.high
    env_dx = (env_high - env_low) / NUM_BUCKETS
    return tuple(((obs - env_low) / env_dx).astype(int))

def save_image(image_array, episode, step):
    """Saves an image file from an array of pixel values."""
    img = Image.fromarray(image_array)
    img.save(f'cartpole_images/episode_{episode}_step_{step}.png')

def run_episode(env, Q, learning_rate, discount, epsilon, episode, render=False):
    current_state = discretize(env.reset())
    done = False
    step = 0

    while not done:
        if render:
            image_array = env.render(mode='rgb_array')
            save_image(image_array, episode, step)

        if np.random.random() < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q[current_state])

        observation, reward, done, info = env.step(action)
        next_state = discretize(observation)
        old_value = Q[current_state + (action,)]
        next_max = np.max(Q[next_state])

        new_value = old_value + learning_rate * (reward + discount * next_max - old_value)
        Q[current_state + (action,)] = new_value

        current_state = next_state
        step += 1

    return step  # Return the number of steps taken

env = gym.make('CartPole-v0')
NUM_BUCKETS = (1, 1, 6, 3)
Q = np.zeros(NUM_BUCKETS + (env.action_space.n,))

learning_rate = 0.1
discount = 0.99
episodes = 500
epsilon = 0.1

# Create a directory to store images
import os
if not os.path.exists('cartpole_images'):
    os.makedirs('cartpole_images')

for episode in range(episodes):
    steps = run_episode(env, Q, learning_rate, discount, epsilon, episode, render=True)
    print(f'Episode {episode} - Steps taken: {steps}')

env.close()
```
